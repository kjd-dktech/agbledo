# Prétraitement et Construction du Dataset


```python
import sys
from pathlib import Path

PROJECT_ROOT = Path(".").resolve().parent if Path(".").resolve().name == "notebooks" else Path(".").resolve()
sys.path.append(str(PROJECT_ROOT))

from src.data.preprocessing import DatasetBuilder
```

```python
builder = DatasetBuilder(project_root=PROJECT_ROOT, random_state=42)

summary_df = builder.build_dataset()
```

### Audit

```python
print(f"Total d'images conservées dans processed : {len(summary_df)}")
display(summary_df.sample(5))
```

```python
# Répartition Train/Val/Test
split_counts = summary_df.groupby(['class', 'split']).size().unstack(fill_value=0)

# Réorganisation des colonnes dans un ordre logique
split_counts = split_counts[['train', 'val', 'test']]
split_counts['total'] = split_counts.sum(axis=1)

# Calcul des pourcentages réels post-stratification
split_counts['train_%'] = (split_counts['train'] / split_counts['total'] * 100).round(1)
split_counts['val_%'] = (split_counts['val'] / split_counts['total'] * 100).round(1)
split_counts['test_%'] = (split_counts['test'] / split_counts['total'] * 100).round(1)

display(split_counts)
```
